```python
import pandas as pd
import numpy as np
```

```python
HeroesofPymoli_file = "../Pandas/HeroesofPymoli_data.csv"
HeroesofPymoli_df = pd.read_csv(HeroesofPymoli_file)

HeroesofPymoli_df.head()
```

```python
Player_Count = len(HeroesofPymoli_df["SN"].unique())
Total_Player_Count = pd.DataFrame({"Total Players": [Player_Count]})

Total_Player_Count
```

```python
Unique_Items = len(HeroesofPymoli_df["Item Name"].unique())
Avg_Purchase_Price = HeroesofPymoli_df["Price"].mean()
Total_Purchases = len(HeroesofPymoli_df["Item Name"])
Total_Revenue = HeroesofPymoli_df["Price"].sum()

Purchasing_Analysis_Total = pd.DataFrame({"Number of Unique Items": [Unique_Items],
                                           "Average Price": [Avg_Purchase_Price],
                                           "Number of Purchases": [Total_Purchases],
                                           "Total Revenue": [Total_Revenue]})

Purchasing_Analysis_Total
```

```python
Duplicates = HeroesofPymoli_df.drop_duplicates(subset ="SN", keep = "first")
Total_Gender = Duplicates["Gender"].count()
Male = Duplicates["Gender"].value_counts()["Male"]
Female = Duplicates["Gender"].value_counts()["Female"]
Other = Total_Gender - Male - Female

Male_Percentage = (Male / Total_Gender) * 100
Female_Percentage = (Female / Total_Gender) * 100
Other_Percentage = (Other / Total_Gender) * 100

Gender_Demographics = pd.DataFrame({"": ["Male", "Female", "Other/Non-Disclosed"],
                    "Total Count": [Male, Female, Other],
              "Percentage of Players": [Male_Percentage, Female_Percentage, Other_Percentage]})

Gender_Demographics
```

```python
Gender_Demographics_df = HeroesofPymoli_df.groupby(["Gender"])

Purchase_Count = Gender_Demographics_df["SN"].count()
Avg_Purchase_Price = (Gender_Demographics_df["Price"].mean()).round(2)
Total_Purchase_Value = Gender_Demographics_df["Price"].sum()

Duplicates = HeroesofPymoli_df.drop_duplicates(subset='SN', keep="first")
Gender_Duplicates = Duplicates.groupby(["Gender"])

Avg_Total_Purchase = (Gender_Demographics_df["Price"].sum() / Gender_Duplicates["SN"].count()).round(2)

Purchasing_Analysis_Gender = pd.DataFrame({"Purchase Count": Purchase_Count,
                                  "Average Purchase Price": Avg_Purchase_Price,
                                  "Total Purchase Value": Total_Purchase_Value,
                                  "Avg Total Purchase per Person": Avg_Total_Purchase})

Purchasing_Analysis_Gender
```

```python
Age_Bins = [0, 10, 15, 20, 25, 30, 35, 40, 120]
Age_Ranges = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

Age_Demographics_df = HeroesofPymoli_df.copy()
Age_Demographics_df["Age Groups"] = pd.cut(Age_Demographics_df["Age"], Age_Bins, labels = Age_Ranges)
Group_Bin = Age_Demographics_df.groupby(["Age Groups"])

Age_Groups_Count = Group_Bin["SN"].count()
Total_Count = HeroesofPymoli_df["SN"].count()
Percentage_of_Players = ((Age_Groups_Count / Total_Count) * 100).round(2)

Age_Demographics = pd.DataFrame({"Total Count": Age_Groups_Count,
                         "Percentage of Players": Percentage_of_Players})

Age_Demographics
```

```python
Age_Bins = [0, 10, 15, 20, 25, 30, 35, 40, 120]
Age_Ranges = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

Age_Analysis_df = HeroesofPymoli_df.copy()
Age_Analysis_df["Age Groups"] = pd.cut(Age_Analysis_df["Age"], Age_Bins, labels = Age_Ranges)
Grouped_Bin = Age_Analysis_df.groupby(["Age Groups"])

Purchase_Count = Grouped_Bin["Age"].count()
Avg_Purchase_Price = Grouped_Bin["Price"].mean()
Total_Purchase_Value = Grouped_Bin["Price"].sum()

Duplicates = HeroesofPymoli_df.drop_duplicates(subset='SN', keep="first")
Duplicates["Age Groups"] = pd.cut(Duplicates["Age"], Age_Bins, labels = Age_Ranges)
Duplicates = Duplicates.groupby(["Age Groups"])

Avg_Total_Purchase = (Grouped_Bin["Price"].sum() / Duplicates["SN"].count())

Age_Analysis = pd.DataFrame({"Purchase Count": Purchase_Count,
                         "Average Purchase Price": Avg_Purchase_Price,
                         "Total Purchase Value": Total_Purchase_Value,
                         "Avg Total Purchase Per Person": Avg_Total_Purchase})

Age_Analysis.head(8)
```

```python
SN = HeroesofPymoli_df.groupby(["SN"])
Purchase_Count = SN["Item ID"].count()
Total_Purchase_Value = SN["Price"].sum()
Avg_Purchase_Price = (Total_Purchase_Value / Purchase_Count)

Top_Spenders = pd.DataFrame({"Purchase Count": Purchase_Count,
                         "Average Purchase Price": Avg_Purchase_Price,
                         "Total Purchase Value": Total_Purchase_Value})

Top_Spenders = Top_Spenders.sort_values("Total Purchase Value", ascending=False) 
Top_Spenders["Average Purchase Price"] = Top_Spenders["Average Purchase Price"].map("${:.2f}".format)
Top_Spenders["Total Purchase Value"] = Top_Spenders["Total Purchase Value"].map("${:.2f}".format)
Top_Spenders = Top_Spenders[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]

Top_Spenders.head()
```

```python
Item_ID_Name = HeroesofPymoli_df.groupby(["Item ID", "Item Name"])
Purchase_Count = Item_ID_Name["SN"].count()
Price_Sum = Item_ID_Name["Price"].sum()
Item_Price = (Price_Sum / Purchase_Count)
Total_Purchase_Value = (Item_Price * Purchase_Count)

Popular_Items = pd.DataFrame({"Purchase Count": Purchase_Count,
                          "Item Price": Item_Price,
                          "Total Purchase Value": Total_Purchase_Value})

Popular_Items = Popular_Items.sort_values("Purchase Count", ascending = False) 
Popular_Items["Item Price"] = Popular_Items["Item Price"].map("${:.2f}".format)
Popular_Items["Total Purchase Value"] = Popular_Items["Total Purchase Value"].map("${:.2f}".format)
Popular_Items = Popular_Items[["Purchase Count", "Item Price", "Total Purchase Value"]]

Popular_Items.head()
```

```python
Item_ID_Name = HeroesofPymoli_df.groupby(["Item ID", "Item Name"])
Purchase_Count = Item_ID_Name["Gender"].count()
Total_Purchase_Value = Item_ID_Name["Price"].sum()
Item_Price = (Total_Purchase_Value / Purchase_Count)

Profitable_Items = pd.DataFrame({"Purchase Count": Purchase_Count,
                              "Item Price": Item_Price,
                              "Total Purchase Value": Total_Purchase_Value})


Profitable_Items = Profitable_Items.sort_values("Total Purchase Value", ascending=False) 
Profitable_Items["Item Price"] = Profitable_Items["Item Price"].map("${:.2f}".format)
Profitable_Items["Total Purchase Value"] = Profitable_Items["Total Purchase Value"].map("${:.2f}".format)
Profitable_Items = Profitable_Items[["Purchase Count", "Item Price", "Total Purchase Value"]]

Profitable_Items.head()
```
